```python
# N * M
# 칸은 육지(0), 바다(1)

import sys
n, m = tuple(map(int, sys.stdin.readline().split()))  # N * M

# 게임 캐릭터의 칸의 좌표(x, y) 바라보는 방향 dir_num
x, y, dir_num = tuple(map(int, sys.stdin.readline().split()))

# 맵의 정보, 바다(1), 육지(0)
graph = [
    list(map(int, sys.stdin.readline().split()))
    for _ in range(n)
]

# 방문 처리
visited = [
    [False] * n
    for _ in range(n)
]

# 북(0), 동(1), 남(2), 서(3) : idx 기준
# 반 시계 방향 (dir_num = (dir_num + 3) % 4)
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]


def in_range(x, y):
    return x >= 0 and x < n and y >= 0 and y < m


def can_go(x, y):
    if not in_range(x, y):
        return False
    if graph[x][y] == 1:
        return False
    if visited[x][y]:
        return False
    return True


cnt = 0  # 방문한 칸의 수
spin_count = 0  # 회전 횟수

while 1:
    # 반시계 방향으로 회전( 왼쪽으로 90도 회전 )
    dir_num = (dir_num + 3) % 4

    # 회전 횟수 증가
    spin_count += 1

    # 반시계 방향으로 회전했을 때 갈 수 있느냐 판단하기 위해서 nx, ny로 가본다.
    nx, ny = x + dx[dir_num], y + dy[dir_num]

    if can_go(nx, ny):
        cnt += 1  # 이동횟수 증가
        visited[nx][ny] = True  # 방문 처리
        spin_count = 0  # 회전 횟수 초기화
        x, y = nx, ny  # 칸 이동

    if spin_count == 4:
        bx, by = x - dx[dir_num], y - dy[dir_num]  # 뒤로 이동할 때

        if graph[bx][by] == 1:  # 뒤가 바다면 종료!
            break
        else:  # 뒤가 육지면 육지로 이동후, 다시 첫 단계부터 시작
            x, y = bx, by  # 육지 이동
            spin_count = 0  # 회전횟수 초기화

print(cnt)
```
