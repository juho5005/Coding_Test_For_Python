### 예제인 00110 을 입력받을 때 
### list(map(int, input().split())) 으로 입력받는것으로 착각했다.
### '00110' 을 문자열로 입력받는것 기억하기..

# DFS 풀이

```python
# N * M 크기의 얼음 틀
# 구멍이 뚫린 부분은 0, 칸막이가 존재하는 부분은 1
# 얼음 틀이 주어졌을 때 상화좌우가 연결되어있는 아이스크림의 총 개수를 구하시오

# DFS 방법 사용

import sys 
n, m = tuple(map(int, sys.stdin.readline().split()))
sys.setrecursionlimit(10**6)

graph = [
    list(sys.stdin.readline().strip())
    for _ in range(n)
]

visited = [
    [False for _ in range(m)]
    for _ in range(n)
]

def in_range(x, y) :
    return x>=0 and x<n and y>=0 and y<m

def can_go(x, y) :
    if not in_range(x, y) :
        return False 
    if visited[x][y] : 
        return False 
    if graph[x][y] == '1' :
        return False 
    return True 

def dfs(x, y) :
    dxs = [-1, 1, 0, 0]
    dys = [0, 0, -1, 1]

    for dx, dy in zip(dxs, dys) :
        nx, ny = x+dx, y+dy 

        if can_go(nx, ny) :
            visited[nx][ny] = True 
            dfs(nx, ny)
            
cnt = 0
for i in range(n) :
    for j in range(m) :
        if not visited[i][j] and graph[i][j] == '0' :
            visited[i][j] = True 
            dfs(i, j)
            cnt += 1

print(cnt)
```

# BFS 풀이

```python
# N * M 크기의 얼음 틀
# 구멍이 뚫린 부분은 0, 칸막이가 존재하는 부분은 1
# 얼음 틀이 주어졌을 때 상화좌우가 연결되어있는 아이스크림의 총 개수를 구하시오

# BFS 방법 사용

from collections import deque 
import sys 
n, m = tuple(map(int, sys.stdin.readline().split()))
q = deque()

dxs = [-1, 1, 0, 0]
dys = [0, 0, -1, 1]

graph = [
    list(sys.stdin.readline().strip())
    for _ in range(n)
]

visited = [
    [False for _ in range(m)]
    for _ in range(n)
]

def in_range(x, y) :
    return x>=0 and x<n and y>=0 and y<m

def can_go(x, y) :
    if not in_range(x, y) :
        return False 
    if visited[x][y] : 
        return False 
    if graph[x][y] == '1' :
        return False 
    return True 

def bfs() :
    while q :
        x, y = q.popleft()
        for dx, dy in zip(dxs, dys) :
            nx, ny = x+dx, y+dy 

            if can_go(nx, ny) :
                visited[nx][ny] = True 
                q.append((nx, ny))

cnt = 0
for i in range(n) :
    for j in range(m) :
        if not visited[i][j] and graph[i][j] == '0' :
            visited[i][j] = True 
            q.append((i, j))
            bfs()
            cnt += 1

print(cnt)
```
